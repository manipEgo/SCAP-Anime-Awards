```python
import pandas as pd

csv_path = "../data/SCAPPAA_2023.csv"

# read csv
df = pd.read_csv(csv_path)
df.columns
```

```python
title_columns = [
    "1、投票列表：请尽量填入完整的中文动画标题。—印象深刻的年度动画（权重4）（必答）",
    "1、想要推荐的精彩动画A（权重2）",
    "1、想要推荐的精彩动画B（权重2）",
    "1、值得一看的优秀动画A（权重1）",
    "1、值得一看的优秀动画B（权重1）",
]
score_columns = [
    "3、更改投票权重（总计10）：—年度动画",
    "3、精彩动画A",
    "3、精彩动画B",
    "3、优秀动画A",
    "3、优秀动画B",
]
comment_columns = [
    "5、动画瓶语：—年度动画",
    "5、精彩动画A",
    "5、精彩动画B",
    "5、优秀动画A",
    "5、优秀动画B",
]

score_dict = {}
comment_dict = {}

# calculate score: total score = sum(score * weight)
for title_column, score_column in zip(title_columns, score_columns):
    for title, score in zip(df[title_column], df[score_column]):
        if score == "(跳过)":
            if title_column == title_columns[0]:
                score = 4
            elif title_column == title_columns[1] or title_column == title_columns[2]:
                score = 2
            else:
                score = 1
        if title in score_dict:
            score_dict[title] += int(score)
        else:
            score_dict[title] = int(score)

# record comments
for title_column, comment_column in zip(title_columns, comment_columns):
    for title, comment in zip(df[title_column], df[comment_column]):
        if comment == "(跳过)" or comment == "(空)":
            continue
        if title in comment_dict:
            comment_dict[title].append(comment)
        else:
            comment_dict[title] = [comment]

# print scores in descending order
# print("Scores:")
# for title, score in sorted(score_dict.items(), key=lambda x: x[1], reverse=True):
#     print(title, score)

# print comments
print("Comments:")
for title, comments in comment_dict.items():
    print(title)
    for comment in comments:
        print(" > ", end="")
        print(comment)
    print("=" * 10)
```
